```python
f500 = pd.read_csv("f500.csv")
f500.head()
```

## Problem : Need the name of "ceo","company","profits" of the most profitable company at every country

```python
bol = f500[f500["country"] == "USA"].sort_values("profits", ascending = False)
ceo_comp_profit= bol[["ceo","company","profits"]]
ceo_comp_profit                                  # separate and deceding the data against profit for USA
```

```python
unique_country = f500["country"].unique()
unique_country                                # find out the unique countries in give data 
```

```python
Top_comp_by_profit = {}       #making dic for top profit company for every country


for i in unique_country:
    bol = f500[f500["country"] == i].sort_values("profits", ascending = False)[["ceo","company","profits"]]
    Top_comp_by_profit[i] = bol
    
Top_comp_by_profit
```

```python
Top_comp_by_profit["USA"].iloc[0]  # indexing for USA top proift compnay 
```

```python
only_Top_comp_prof = {}         # indexing for top profitable company for every country 

for i in Top_comp_by_profit:
    Top_comp_every_country =Top_comp_by_profit[i].iloc[0]
    only_Top_comp_prof[i] = Top_comp_every_country
```

```python
only_Top_comp_prof
```

```python
df_Top_comp_by_ptofit = pd.DataFrame(only_Top_comp_prof).T   # converting dic into data frame for good undersating 
df_Top_comp_by_ptofit.sort_values("profits", ascending = False)  #also sort the data according to profit for future analysis 
```
